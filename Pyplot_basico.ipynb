#Pyplot b√°sico

```python
#Importamos la libreria
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Hacemos que genere 11 elementos del 1 al 5.
# Lo almacenamos en una variable.
# Creamos otra variable y donde almacene los elementos que creamos elevado al cuadrado.

x = np.linspace(0,5,11)
y = x**3
x
```

```python
# Visualizacion de una grafica sencilla
plt.plot(x,y)

# Le decimos que nos muestre esa grafica
plt.show()
```

```python
#Otro ejemplo
#Llamamos la funcion de la grafica, que es plot. Le pasamos las variables las cuales tiene los numeros.
# Despues le podemos especificar el color.
plt.plot(x,y,"orange")
plt.show
```

```python
#Otro ejemplo
#Podemos crear una viz con una grafica pero con una figura distinta.
#Pasamos la libreria, luego la funcion, y las variables con el valor.
#Luego con una letra podemos decirle el color y la figura.

plt.plot(x,y, "b x")
```

```python
#Otro ejemplo
#Podemos crear una viz con una grafica pero con una figura distinta.
#Pasamos la libreria, luego la funcion, y las variables con el valor.
#Luego con una letra podemos decirle el color y la figura.
#Podemos decirle que le de cuntinuida a la linea con s-

plt.plot(x,y, "bs-")
```

```python
#Otro ejemplo
#Podemos crear una viz con una grafica pero con una figura distinta.
#Pasamos la libreria, luego la funcion, y las variables con el valor.
# O unos peque√±os puntos y un diamante con D y los puntos con :

plt.plot(x,y, "yD:")
```

```python
# Otro ejemplo pero en histograma
# Usamos la libreria. Le ponemos la terminacion hist.
plt.hist(x, y)
plt.show()
```

```python
# Otro ejemplo de grafica de pastel
# pasamos la libreria.
# le pasamos la funcion de la libreria.
# Le pasamos la variables.
#Podemos usar la funcion de show para mostrar la grafica.
plt.pie(x)
plt.show()
```

```python
# Otro ejemplo con un grafico de scatter
# Traemos la libreria.
# Usamos la funcion.
# Le pasamos la variable.
# Mostramos en pantalla la funcion

plt.scatter(x, y)
plt.show()
```

```python
# Otro ejemplo con un grafico de boxplot
# Traemos la libreria.
# Usamos la funcion.
# Le pasamos la variable.
# Mostramos en pantalla la funcion

plt.boxplot(x)
plt.show()
```
