## Parametros mas usuados en seaborn

```python
# Importamos las librerias
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargamos unso datos predeterminados que estan en seaborn
# Es un data set de las propinas de un restaurante
tips = sns.load_dataset("tips")
tips
```

## Parametros basicos

```python
# Displot
# usamos la libreria sns
# displot es la funcion para hacer el grafico.
# data = tip es el data set
# x es el primer parametros
# y es el segundo parametro
sns.displot(data=tips, x="total_bill")
```

```python
# Displot
# usamos la libreria sns
# displot es la funcion para hacer el grafico.
# data = tip es el data set
# x es el primer parametros
# y es el segundo parametro
#hue hace un agrupamiento
sns.displot(data=tips, x="total_bill", y="tip",hue="sex")
```

```python
# Displot
# usamos la libreria sns
# displot es la funcion para hacer el grafico.
# data = tip es el data set
# x es el primer parametros
# y es el segundo parametro
#hue hace un agrupamiento
# funciona tambien cuando solo hay un histograma.
sns.displot(data=tips, x="total_bill",hue="sex")
```

```python
# Displot
# usamos la libreria sns
# displot es la funcion para hacer el grafico.
# data = tip es el data set
# x es el primer parametros
# y es el segundo parametro
#hue hace un agrupamiento
# funciona tambien cuando solo hay un histograma.
#Con kind podemos definir que tipo de grafica queremos.
#Legend para la leyende
#paleta de colores con palette
# alpha para darle transparencia
sns.displot(data=tips, x="total_bill",hue="sex", kind="kde", legend=True, palette="dark", alpha = 0.12)
```
