#Leyendas, etiquetas, títulos, tamaño


```python
#Importamos las librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
#creamos la lista de valores con linspace
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# Creamos nuestros graficos

# Lienzo --
# escribimos nuestra variable.
# usamos axes para el grafico.
# plt para la libreria y subplot para el grafico en si.
# Entre los parentesis ponemos 1 fila y dos columnas.
#Cambiar el tamaño del lienzo con figsize = (altura, anchura)
fig, axes = plt.subplots(1, 2, figsize=(10,5))

# Grficos --
# usamos el axes y por su numero de posicion lo manejamos, en esta caso es el 0.
# le pasamos la funcion plot y las variables con su valores.
#Le podemos dar diseño a la grafica con labels
# Tambien una leyenda con legend
axes[0].plot(x,y, "b", label="linspace(x)")
axes[0].legend()


#Grafico dos --
#Usamos el axes de por su numero que es 1 lo modificamos.
# Le pasamos la funcion plot y sus las varianles que vamos a suar
axes[1].plot(y,x, "r", label="$sin(y)$")
axes[1].legend()

#Titulos --
# Grafica 0
# Para crear un titulo usamos la funcion set con el comando title
# tambien axes y la ubicacion y la etiqueta title. Y entre comillas el titulo.
axes[0].set_title("Relacion entre X & Y")

# Grafica 1
# Se usa el axes, la posicion, el set_title y entre parentesis y comillas el titulo.
axes[1].set_title("Relacion entre Y & x")

# Label --
# Podemos crear un label tambien.
# lo mismo: # Se usa el axes, la posicion, el set_label (antes del label la posicon)y entre parentesis y comillas el titulo.
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")


#label 1
axes[1].set_xlabel("y")
axes[1].set_ylabel("y")
```
