## Relation

```python
#importamos las librerias
import seaborn as sns
import matplotlib.pyplot as plt

# dataset
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
#grafica
#usamos la libreria
# la funcion scatterplot
# entre parentesis la base de datos y las variables
#segmentamos por dia con hue
#style para decirle el time o el horario
# le podemos dar el tamaño de los scatter con size
#marker para el diseño del scatter
markers = {"Lunch":"D","Dinner":"s"}
sns.scatterplot(data=tips,x="total_bill", y="tip", hue="day", style="time", size="size", markers = markers)

#leyenda
#con legend
#loc para el centro
plt.legend(loc="center",bbox_to_anchor=(1.12,0.6))

#lo mostramos en pantalla
plt.show
```

```python
# Otra grafica
#lineplot ayuda a ver la relacion entre las dos variables
sns.lineplot(data=tips,x="total_bill", y="tip", hue="day", style="time", size="size", markers = markers)
plt.show()
```
