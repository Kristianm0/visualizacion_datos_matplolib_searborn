#Colores y estilos


```python
#importamos la libreria
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creamos los valores
x = np.linspace(0,5,11)
```

```python
# mostramos en pantalla la funcion de style.avaible
#style.available muestra los estilos por defauts.
print(plt.style.available)
```

```python
# Dise√±o del grafico
# plt usamos la libreria
# style, usamos la funcion estilo y usamos el comando use para usarlo
# entre los parentesis le damos el tema
plt.style.use("classic")
```

```python
# Hacemos nuestro lienzo o figura
# fig par ala variable ax con el comando,
# plt como la libreria, subplots como la la funcion
# figsize como la funcion que crear la medida
fig, ax = plt.subplots(figsize=(8,8))

# Creamos los valores de la grafica
ax.plot(x,x+1, "red")
ax.plot(x,x+2, "b")
ax.plot(x,x+9, "yellow")
ax.plot(x,x+6, "orange")
```

## Modificaciones matlab

```python
# Hacemos nuestro lienzo o figura
# fig par ala variable ax con el comando,
# plt como la libreria, subplots como la la funcion
# figsize como la funcion que crear la medida
fig, ax = plt.subplots(figsize=(8,8))

# Creamos los valores de la grafica
#Podemos cambiar el estilo de la linea, poniendi entre comillas
# 1)  --: Linea continua
# 2) - : Linea normal
# 3) .- : Puntos y lineas
ax.plot(x,x+1, 'r--')
ax.plot(x,x+2, "b-")
ax.plot(x,x+9, "y.-")
```

## Con pyplot

```python
# Hacemos nuestro lienzo o figura
# fig par ala variable ax con el comando,
# plt como la libreria, subplots como la la funcion
# figsize como la funcion que crear la medida
fig, ax = plt.subplots(figsize=(8,8))

# Creamos los valores de la grafica
#Podemos modificar el estilo con .color
# tambien recibe colores rgb
# el nivel de tranparencia
# con linewidth se puede configurar el grosor de la linea
#linestyle el estilo de la linea
# Marcadores y su tama√±o con marker y markesize
# cambiar color de marker
ax.plot(x,x+1,color = "green", alpha = 0.5, linewidth=8, linestyle = "-", marker = "o", markersize=8 )
ax.plot(x,x+2, color = "blue", linewidth = 10, linestyle = "--", marker = "x", markerfacecolor = "b")
ax.plot(x,x+9, color = "#882469", linewidth = 6, linestyle = "dashed", marker = "v")
```
