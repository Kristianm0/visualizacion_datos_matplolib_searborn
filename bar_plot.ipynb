## Bar plot

```python
#importamos las librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creamos una lista para el ejemplo
country = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", "GERMANY"]
population = [1000, 11110, 2000, 40303, 20020]
```

```python
# Creamos el grafico de barra
# usamos la libreria, la funcion bar, y entre los parentesis las variables
plt.bar(country,population)

# Le decimo que nos muestre la grafica
plt.show()
```

```python
# Creamos el grafico de barra
# usamos la libreria, la funcion bar, y entre los parentesis las variables
# con width le damos el ancho
# con color se modifica su color
plt.bar(country,population, width = 0.5, color = ["red", "blue", "yellow", "green"])

# Le decimo que nos muestre la grafica
plt.show()
```

```python
#Podemos cambiar el nombre de alguna variable
# Creamos el grafico de barra
# usamos la libreria, la funcion bar, y entre los parentesis las variables
# con width le damos el ancho
# con color se modifica su color
# con rotation = 45 podemos rotar el nombre
plt.bar(country,population, width = 0.5, color = ["red", "blue", "yellow", "green"])

#Cambiar el nombre con xticks y arange, entre los parentesis el numero de donde esta el valor
plt.xticks(np.arange(5), ("India", "Japon", "Mexico", "Colombia", "Alemania"), rotation=45)

# Le decimo que nos muestre la grafica
plt.show()
```

```python
#Para tener las barra pero de manera horizontal
#la libreria, la funcion barh y las variables entre los parentesis.
plt.barh(country, population)
plt.show()
```
