#Metodo orientado a objetos

```python
#importamos las librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creamos algunas variables
# Creamos los valores con la funcion linspace

x = np.linspace(0,5,11)
y = x**2
```

```python
# Metodo orientado a objetos

#Creamos nuestro lienzo con la funcion figure y lo almacenamos en una variable.
fig = plt.figure()

#creamos las graficas que se llamaran axes, al objeto de le añade un axes
# Entre los parentesis y corchetes le damos los valores de los indices.
#  Los primeros dos valores dentro de los corchetes son: de la figura en el lienzo. Los otros dos son del: tamaño de ese axes o figura.
axes = fig.add_axes([0.1,0.1,0.8,0.9])

#creamos un plot de nuestras variables
axes.plot(x,y,"b")
# La mostramos en pantalla
fig.show()
```

```python
# Otro ejemplo
# Metodo orientado a objetos

#lienzo--
#Creamos nuestro lienzo con la funcion figure y lo almacenamos en una variable.
fig = plt.figure()

#valores de lienzo ---
#creamos las graficas que se llamaran axes, al objeto de le añade un axes
# Entre los parentesis y corchetes le damos los valores de los indices.
#  Los primeros dos valores dentro de los corchetes son: de la figura en el lienzo. Los otros dos son del: tamaño de ese axes o figura.
axes = fig.add_axes([0.1,0.1,0.8,0.9])

#diseño --
# creamos otra grafica axes_2
axes_2 = fig.add_axes([0.2,0.55,0.4,0.3])

# valores de variables ---
#creamos un plot de nuestras variables
axes.plot(x,y,"b")

#Le damos diseño al axes_2
# grafica
axes_2.plot(x,y,"r")
#lienzo
axes_2.set_facecolor("gray")

# La mostramos en pantalla
fig.show()
```
