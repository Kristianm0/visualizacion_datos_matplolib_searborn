## Subplot con objetos

```python
#Importamos las librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creamos las variables con los valores
x = np.linspace(0,5,11)

# Le aplicamos una funcion de seno a la variable y lo guardamos en y
y = np.sin(x)
```

```python
# definomos nuestros objetos
fig, axes = plt.subplots()
axes.plot(x,y,"b")
```

```python
# definomos nuestros objetos
fig, axes = plt.subplots(nrows=1, ncols=2)
# Los corchetes que estan antes del .plot es la posicion donde estara la figura.
#grafico 1
axes[0].plot(x,y,"b")
# grafico 2
axes[1].plot(y,x,"r")
```

```python
#otro ejemplo definiendo los axes
# definomos nuestros objetos, y accedemos a ellos definiendolos
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

# Los corchetes que estan antes del .plot es la posicion donde estara la figura.

#grafico 1
ax1.plot(x,y,"b")
# grafico 2
ax2.plot(y,x,"r")
```

```python
#otro ejemplo con 4 graficas
# definomos nuestros objetos, y accedemos a ellos definiendolos
fig, axes = plt.subplots(2, 4)

# Para acceder
#grafico 1
# le decimos que en la posicion 0,0 del lienzo, un plot de x conseno x.
axes[0,0].plot(x,np.cos(x), "orange")

# posicion 2 coseno elevado al cuadrado
axes[0,1].plot(x,np.cos(x)**2)


# Para el grafico dos le aplicamos la funcion tangente
axes[0,2].plot(x,np.tan(x), "black")

# Para la posicion 3 le aplicamos seno
axes[0,3].plot(x,np.sin(x), "red")

# grafico 2
#ax2.plot(y,x,"r")
```

```python
#otro ejemplo con 4 graficas definiendo las graficas desde el principio
# definomos nuestros objetos, y accedemos a ellos definiendolos
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)

# Para acceder
#grafico 1
# le decimos que en la posicion 0,0 del lienzo, un plot de x conseno x.
ax1.plot(x,np.cos(x), "orange")

# posicion 2 coseno elevado al cuadrado
ax2.plot(x,np.cos(x)**2)


# Para el grafico dos le aplicamos la funcion tangente
ax3.plot(x,np.tan(x), "black")

# Para la posicion 3 le aplicamos seno
ax4.plot(x,np.sin(x), "red")



# Podemos decirle que le de espacio a las graficas
fig.tight_layout()
```
