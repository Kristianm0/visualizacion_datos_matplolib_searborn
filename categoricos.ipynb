```python
#importamos las librerias
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# creamos los datos
# creamos la variable para almacenar los datos
# la libreria sns y la funcion load_dataset
#tips que es el datasete por defautl
tips = sns.load_dataset("tips")

#lo mostramos en pantalla
tips.head(2)
```

```python
#grafica
# usamos la libreria
#funcion de countplot para la grafica
# entre los parentesis los valores y para el eje x
# hue para segmentar por sexo.

sns.countplot(data=tips,x="day",hue="sex")

#lo mostramos en pantalla
plt.show()
```

```python
#grafica
# usamos la libreria
#funcion de striplot para la grafica
# entre los parentesis los valores y para el eje x
# hue para segmentar por sexo.
# lo graficamos otra variable en el eje y
# usamos dodge para separar las variables

#podemos modificar la grafica con figure
plt.figure(figsize=(6,6))

sns.stripplot(data=tips,x="day", y="total_bill", hue="sex", dodge=True)

#lo mostramos en pantalla
plt.show()
```

```python
#grafica
# usamos la libreria
#funcion de swarmplot para la grafica
# entre los parentesis los valores y para el eje x
# hue para segmentar por sexo.
# lo graficamos otra variable en el eje y
# usamos dodge para separar las variables

#podemos modificar la grafica con figure
plt.figure(figsize=(6,6))

sns.swarmplot(data=tips,x="day", y="total_bill", hue="sex", dodge=True)

#lo mostramos en pantalla
plt.show()
```

```python
#grafica
# usamos la libreria
#funcion de boxplot para la grafica
# entre los parentesis los valores y para el eje x
# hue para segmentar por sexo.
# lo graficamos otra variable en el eje y
# usamos dodge para separar las variables

#podemos modificar la grafica con figure
plt.figure(figsize=(6,6))

sns.boxplot(data=tips,x="day", y="total_bill", hue="sex", dodge=True)

#lo mostramos en pantalla
plt.show()
```

```python
#grafica combinadas
# usamos la libreria
#funcion de boxplot para la grafica
# entre los parentesis los valores y para el eje x
# hue para segmentar por sexo.
# lo graficamos otra variable en el eje y
# usamos dodge para separar las variables

#podemos modificar la grafica con figure
plt.figure(figsize=(6,6))

sns.boxplot(data=tips,x="day", y="total_bill", hue="sex", dodge=True)
sns.swarmplot(data=tips,x="day", y="total_bill", hue="sex", dodge=True, color = "0" )

#lo mostramos en pantalla
plt.show()
```

```python
#grafica combinadas
# usamos la libreria
#funcion de catplot para la grafica
# entre los parentesis los valores y para el eje x
# hue para segmentar por sexo.
# lo graficamos otra variable en el eje y
# usamos dodge para separar las variables
# col es y time es un parametro que puede indicar la hora

#podemos modificar la grafica con figure
plt.figure(figsize=(6,6))

sns.catplot(data=tips,x="day", y="total_bill", hue="sex", dodge=True, col="time")

#lo mostramos en pantalla
plt.show()
```
