## Otras graficas

```python
#Importamos las librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creamos base de datos
# data como nuestra variable
# np como nuestra libreria, random como la funcion.
# randit como otra funcion y entre parentesis los numeros que quiero, en 1 en 1, hasta el 50 y que sean 100.

data = np.random.randint(1,50,100)
```

```python
# Histograma --
# Usamos la libreria plt
# Usamos la funcion del histograma hist
# Bins para el numero de frecuencia de un grafico
# Con width el ancho de los graficos
#histtype para el tipo de barra el cual es bar
hist = plt.hist(data, bins=50, width=0.4, histtype = "bar")
hist
```

##histograma

```python
# Histograma -- barra step
# Usamos la libreria plt
# Usamos la funcion del histograma hist
# Bins para el numero de frecuencia de un grafico
# Con width el ancho de los graficos
#histtype para el tipo de barra el cual es step
hist = plt.hist(data, bins=50, histtype = "step")
hist
```

## box plot

```python
# Grafica de box plot
# usamos la libreria plt.
# la funcion boxplot, y la variable entre parentesis.

boxplot = plt.boxplot(data)
```

```python
# Grafica de box plot
# usamos la libreria plt.
# la funcion boxplot, y la variable entre parentesis.
# con verl podemos poner el grafico vertical
# patch_artist para darle color

boxplot = plt.boxplot(data, vert=False, patch_artist=True)
```

```python
# Grafica de box plot
# usamos la libreria plt.
# la funcion boxplot, y la variable entre parentesis.
# con verl podemos poner el grafico vertical
# patch_artist para darle color
# notch centra los datos a la mediana


#Modifiquemos la variable
# Con .append agregamos un nuevo dato
data = np.append(data,200)
boxplot = plt.boxplot(data, vert=False, patch_artist=True, notch=True)
```

```python
# Grafica de box plot
# usamos la libreria plt.
# la funcion boxplot, y la variable entre parentesis.
# con verl podemos poner el grafico vertical
# patch_artist para darle color
# notch centra los datos a la mediana


#Modifiquemos la variable
# Con .append agregamos un nuevo dato
# Podemos excluir donde no hayan datos
data = np.append(data,200)
boxplot = plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers = False)
```

## Scatter

```python
# Creamos algunas variables
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30* np.random.rand(n))**2
colors = np.random.rand(n)
```

```python
# scatter
#veamos la relacion entre dos distintas variables
# usamos la libreria plt, luego la funcion scatter y dentro de los parentesis la variable.
plt.scatter(x,y)
```

```python
# scatter
#veamos la relacion entre dos distintas variables
# usamos la libreria plt, luego la funcion scatter y dentro de los parentesis la variable.
# Que grafique el area con s=area, que area es una variable que creamos
# Puedo poner la variable color que creamos
# La figura tambien se puede cambiar
# alpha para la transparencia
plt.scatter(x,y, s=area, c= colors, marker="p", alpha=0.5)
```
