##Distribuciones

```python
# Importamos la librerias
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# creamos una dataset
# almacenamos el dataset en una variable.
# usamos la libreria con la funcion load_dataset
# entre parentesis el data set por defaut que tiene seaborn
tips = sns.load_dataset("tips")

# traemos la cabezera del dataset
tips.head()
```

```python
# Aplicamos un grafico
# pasamos la libreria sns
#pasamos la funcion histplot.
# le especificamos en dataset
# le decimos los valores de los ejes
# bins para agrupar las barras
sns.histplot(data=tips, x="tip", bins=15 )

#lo mostramos en pantalla
plt.show()
```

```python
# Aplicamos un grafico
# pasamos la libreria sns
#pasamos la funcion histplot.
# le especificamos en dataset
# le decimos los valores de los ejes
# bins para agrupar las barras
# cumulative para acumular los valores
sns.histplot(data=tips, x="tip", bins=15, cumulative = True)

#lo mostramos en pantalla
plt.show()
```

```python
# Aplicamos un grafico
# pasamos la libreria sns
#pasamos la funcion histplot.
# le especificamos en dataset
# le decimos los valores de los ejes
# bins para agrupar las barras
# cumulative para acumular los valores
#hue para agrupar
sns.histplot(data=tips, x="tip", bins=15, cumulative = False, hue="sex")

#lo mostramos en pantalla
plt.show()
```

```python
# Aplicamos un grafico
# pasamos la libreria sns
#pasamos la funcion histplot.
# le especificamos en dataset
# le decimos los valores de los ejes
# bins para agrupar las barras
# cumulative para acumular los valores
#hue para agrupar
# stat alguna funcion de estadistica; frecuency, probability, percent, count,
# multiple para decirlo donde comienza el numero; hay stack, layer, dodge, fill
sns.histplot(data=tips, x="tip", bins=15, cumulative = False, hue="sex", stat="frequency", multiple="dodge")

#lo mostramos en pantalla
plt.show()
```

```python
#Diagrama de densidad
# usamos nuestra libreria.
# la funcion kdeplot para el diagrama de densidad.
# especificamos el dataset, le decimos los valores el eje x
# agrupamos los datos con hue sex
# le pasamos el cumulative
#shade para rellenar el area de abajo de la linea
#bw_adjust para ajustar la grafica de los ejes
sns.kdeplot(data=tips, x="tip", hue="sex", cumulative=False, shade = True, bw_adjust=1)

#mostramos en pantalla la grafica
plt.show()
```

```python
# cremos otra grafica
#pasamos la libreria
# luego la funcion ecdplot
# luego los datos y las variables de eje x
#sat para cambiar el tipo y count para contar cada uno

sns.ecdfplot(data=tips, x="tip", hue="sex", stat="count")

#lo mostramos en pantalla
plt.show()
```

```python
# cremos otra grafica de dispersion
#pasamos la libreria
# luego la funcion displot para la dispersion
# luego los datos y las variables de eje x
#sat para cambiar el tipo y count para contar cada uno
#darle el tipo con kind
# multiple para el tipo distribucion

sns.displot(data=tips, x="tip", hue="sex", stat="count", kind="hist", multiple="stack")

#lo mostramos en pantalla
plt.show()
```
